```python
import json

def create_or_load_dictionary(file_name):
    try:
        with open(file_name, "r") as file:
            data = json.load(file)
    except FileNotFoundError:
        # If the file doesn't exist, create an empty dictionary
        data = {}

    return data

def save_dictionary(file_name, dictionary):
    with open(file_name, "w") as file:
        json.dump(dictionary, file)

def add_word_meaning(dictionary, word, meaning):
    dictionary[word] = meaning

def search_word_meaning(dictionary, word):
    return dictionary.get(word, "Word not found in the dictionary.")

def remove_word_meaning(dictionary, word):
    if word in dictionary:
        del dictionary[word]

def main():
    file_name = "dictionary.json"
    dictionary = create_or_load_dictionary(file_name)

    while True:
        print("\nDictionary Console Tool:")
        print("1. Add Word and Meaning")
        print("2. Search for Meaning")
        print("3. Remove Word and Meaning")
        print("4. Exit")

        choice = input("Enter your choice (1/2/3/4): ")

        if choice == "1":
            word = input("Enter the word: ")
            meaning = input("Enter the meaning: ")
            add_word_meaning(dictionary, word, meaning)
            save_dictionary(file_name, dictionary)
            print("Word and meaning added to the dictionary.")
        elif choice == "2":
            word = input("Enter the word to search: ")
            meaning = search_word_meaning(dictionary, word)
            print(f"Meaning: {meaning}")
        elif choice == "3":
            word = input("Enter the word to remove: ")
            remove_word_meaning(dictionary, word)
            save_dictionary(file_name, dictionary)
            print("Word and meaning removed from the dictionary.")
        elif choice == "4":
            break
        else:
            print("Invalid choice. Please select a valid option.")

if __name__ == "__main__":
    main()
```
